````python
import pandas as pd
import matplotlib.pyplot as plt

# Assume the dataset is named "Toronto_Public_Health_Data.csv"
data = pd.read_csv(r'C:\Users\Beast\visualization\02_activities\assignments\Toronto_Public_Health_Data.csv')

# Clean and analyze the data, assuming we are analyzing the number of cases by year
yearly_cases = data.groupby('year')['cases'].sum()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(yearly_cases.index, yearly_cases.values, marker='o')
plt.title('Yearly Cases of Public Health Issue in Toronto')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.tight_layout()
plt.savefig('visualization_toronto_public_health.png')  # Save as a PNG image
plt.show()
````

````python
# Load the dataset
# 
data = pd.read_csv('Municipalities _dataset.csv')

# Example 1: Line Plot
plt.figure(figsize=(10, 6))
x = data['Time']
y = data['Value']
plt.plot(x, y, color='blue', linestyle='-', label='Value over Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Line Plot of Value over Time')
plt.legend()
plt.grid(True)
plt.show()

# Example 2: Bar Chart
categories = data['Category'].value_counts()
plt.figure(figsize=(10, 6))
categories.plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Bar Chart of Categories')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Example 3: Subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Plot 1: Histogram
axs[0].hist(data['Value'], bins=20, color='green', edgecolor='black')
axs[0].set_title('Histogram of Values')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')
````

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.5, 100)
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Noisy Sine Wave', color='b', linestyle='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Exploring Matplotlib: Noisy Sine Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
**Credit: Data Visualization as the Tip of the Iceberg**

- **D'Ignazio and Klein (2020)** relate the idea of underwater labor to data visualization, emphasizing the hidden work that goes into creating visualizations.

- **Unseen Contributions**: Many contributions that are not immediately visible are essential for a data visualization to exist in its final form. Some of these include:
  - **Community Organizers**: Facilitated data collection.
  - **Designers**: Created color palettes for visualizations.
  - **Technical Writers**: Created alt-text and image descriptions to make visualizations accessible.
  - **Student Research Assistants**: Recorded the data used in visualizations.
  - **IT Support Staff**: Helped host the visualization online.
  - **Caregivers**: Supported team members by taking care of children during the project.

**Notes**: This class focused on acknowledging the often invisible contributions that make data visualizations possible. It is crucial to recognize the collective effort behind every visualization, from data collection to technical support and caregiving, as they all contribute to the final output.
```

```python
import plotly.graph_objects as go
import numpy as np

# Data
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])

# Create the figure
graph = go.Figure()

# Add a trace to the figure
graph.add_trace(go.Scatter(
    x=x1,
    y=y1,
    mode='markers',  # we want points for a scatter plot
    marker=dict(
        size=15,  # point size
        color='hotpink',  # point colour
        opacity=1,  # point transparency/alpha
        line=dict(width=5, color='purple')  # point outline
    )
))

# Update layout
graph.update_layout(
    title='Interactive Pirate Plot',
    xaxis_title='Pirates',
    yaxis_title='Scores',
    width=500,
    height=500
)

# Show the plot
graph.show()
```

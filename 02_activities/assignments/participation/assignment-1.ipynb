
## Class 2: Exploring Matplotlib, Reproducible Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.5, 100)
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Noisy Sine Wave', color='b', linestyle='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Exploring Matplotlib: Noisy Sine Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
--**Notes**: This class focused on reproducibility and how to create more advanced plots with Matplotlib. We used random seeds to ensure that our results are consistent each time the code is run, making our visualizations reproducible.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data
t = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(t)
y2 = np.cos(t)
```

```python
# Customize plot with different styles
plt.plot(t, y1, label='Sine Wave', color='r', linestyle='-')
plt.plot(t, y2, label='Cosine Wave', color='g', linestyle=':')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.title('Customizing Plots: Sine and Cosine Waves')
plt.legend()
plt.grid(True)
plt.show()
```

--**Notes**: The focus of this class was on choosing the appropriate visualization for different types of data. We created a bar chart to represent categorical data and discussed other types of visualizations that suit various data characteristics

## Class 5: Subplots and Combining Visualizations + Accessible Data Visualization


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating subplots to compare different visualizations
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Plot 1: Sine wave
axs[0].plot(x, y1, color='b')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Amplitude')
```

```python
# Plot 2: Cosine wave
axs[1].plot(x, y2, color='r')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
# Plot 2: Cosine wave
axs[1].plot(x, y2, color='r')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

**Notes**: In this class, we explored how to create subplots to present multiple visualizations side-by-side. We also discussed making visualizations more accessible by ensuring readability, appropriate color choices, and clear labeling
